```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Makanan =  ('Donat', 'Pastel', 'Kue Cokelat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberi', 'Puding Cokelat', 'Ice Cream Nutela', 'Kue Cokelat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
Harga = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'Makanan':Makanan, 'Harga':Harga})
df.sort_values(by='Harga',inplace=True, ascending=False)

kor_x = np.arange(len(df))
warna=['#0000FF' for _ in range(len(df))]
warna [-3] = '#FF0000'
warna [-2] = '#FF0000'
warna [-1] = '#FF0000'
plt.bar(kor_x, df['Harga'], tick_label=df['Makanan'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Harga (ribu)')
plt.title('Harga Makanan Penutup')
plt.show
```
