```python
import matplotlib.pyplot as plt

rasa = ('cokelat', 'strawberry', 'vanila')
data = (12, 15 ,3)
warna = ('#D2691E', '#FF0000', '#DA7B06')
highlight = (0, 0, 0.3)

plt.title('Penjualan es krim yang paling diminati')
plt.pie(data, labels = rasa, autopct = '%1.1f%%', colors = warna, explode=highlight, shadow=True)
plt.show
```
